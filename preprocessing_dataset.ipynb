La feature "Time" nel formato HH:MM può essere molto utile per vedere quanti incidenti ci sono per fasce orarie ad esempio.
Decido di creare un'altra colonna conservando quella originale, e nella nuova colonna avrò solo le ore con la regola che superate le .30 siamo nella fascia oraria successiva (es. 23.45 => 00)

```python
import numpy as np
dataset_car.insert(dataset_car.columns.get_loc('Time') + 1, 'Time_Hour', dataset_car['Time'].str.split(':').apply(lambda x: x[0] if int(x[1]) <= 30 else str(int(x[0]) + 1)))
dataset_car['Time_Hour'] = dataset_car['Time_Hour'].replace(['0', '24'], '00')
```
